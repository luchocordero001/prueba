# Descarga ENDES desde INEI en Google Colab

Este notebook prueba el acceso con `curl -I` y descarga el ZIP hacia una carpeta de Google Drive que definas.


```python
# Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Configura la URL y la carpeta de destino en Drive
URL = 'https://proyectos.inei.gob.pe/iinei/srienaho/descarga/SPSS/968-Modulo1629.zip'
OUTPUT_DIR = '/content/drive/MyDrive/ENDES'  # Cambia esta ruta
```

```python
# Verificar cabeceras con curl -I
!curl -I "{URL}"
```

```python
import os
from pathlib import Path
from urllib.request import urlretrieve

Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
filename = URL.split('/')[-1]
output_path = os.path.join(OUTPUT_DIR, filename)

print(f'Descargando a: {output_path}')
urlretrieve(URL, output_path)
print('Descarga completa.')
```
